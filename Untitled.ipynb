```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# קרא את קובץ ה-FCA שלך
df = pd.read_csv("field_convergence_A_20250815_162920/fca_summary.csv")

# נניח שהעמודות הן field_i, field_j, fca_slope
# אם לא – עדכני את השמות כאן
pivot = df.pivot(index="field_i", columns="field_j", values="FCA_slope")

# נשלים סימטריה אם צריך (שדה i↔j)
pivot = pivot.combine_first(pivot.T)

plt.figure(figsize=(14, 12))
sns.heatmap(
    pivot,
    cmap="coolwarm",      # כחול לשלילי, אדום לחיובי
    center=0,             # כך שהצבע האמצעי הוא אפס
    cbar_kws={"label": "FCA slope"}
)
plt.title("Field Convergence Acceleration (FCA) across field pairs", fontsize=16)
plt.xlabel("Field j")
plt.ylabel("Field i")
plt.tight_layout()

plt.savefig("fca_heatmap_pairs.png", dpi=300)
plt.show()
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

"""
Make all paper figures from Parts A–E outputs.

Figures produced (saved under paper_figures_<timestamp>/):
  1) fca_heatmap_pairs.png                      — Heatmap של FCA לכל הזוגות (A)
  2) pair_trend_<i>__<j>.png (top 6)            — Time trends לזוגות ה-FCA המובילים (A)
  3) bes_distribution.png                        — היסטוגרמה של BES לכל החוקרים (B)
  4) scatter_fca_vs_bridge_share.png            — פיזור: FCA מול שיעור גשר (C)
  5) box_bridge_share_top_vs_bottom.png         — Boxplot: שיעור גשרים Top/Bottom FCA (C)
  6) box_cpr_top_vs_bottom.png                  — Boxplot: CPR Top/Bottom FCA (C; מחשב מחדש אם צריך)
  7) pvals_histogram_E.png                      — היסטוגרמת p/q ממודל האפס (E)
"""

import os, json, itertools
from datetime import datetime
from collections import defaultdict, Counter

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ============ Helpers ============

def find_latest_dir(prefix: str):
    cands = [d for d in os.listdir(".") if d.startswith(prefix) and os.path.isdir(d)]
    if not cands: return None
    cands.sort(reverse=True)
    return cands[0]

def ensure_outdir():
    ts = datetime.now().strftime("%Y%m%d_%H%M%S")
    out_dir = f"paper_figures_{ts}"
    os.makedirs(out_dir, exist_ok=True)
    return out_dir

def safe_pair(a,b):
    return tuple(sorted((str(a), str(b))))

# ============ FIGURE 1: FCA heatmap (A) ============

def plot_fca_heatmap(out_dir):
    a_dir = find_latest_dir("field_convergence_A_")
    if not a_dir:
        print("⚠️ Part A output not found; skipping heatmap.")
        return None
    fpath = os.path.join(a_dir, "fca_summary.csv")
    if not os.path.exists(fpath):
        print("⚠️ fca_summary.csv not found; skipping heatmap.")
        return None

    df = pd.read_csv(fpath)
    # pivot על i×j
    pivot = df.pivot(index="field_i", columns="field_j", values="FCA_slope")
    # סימטריה (ליתר ביטחון)
    pivot = pivot.combine_first(pivot.T)

    plt.figure(figsize=(14, 12))
    im = plt.imshow(pivot.values, aspect="auto", cmap="coolwarm", vmin=-np.nanmax(abs(pivot.values)), vmax=np.nanmax(abs(pivot.values)))
    plt.colorbar(im, label="FCA slope")
    plt.xticks(range(len(pivot.columns)), pivot.columns, rotation=90, fontsize=8)
    plt.yticks(range(len(pivot.index)), pivot.index, fontsize=8)
    plt.title("Field Convergence Acceleration (FCA) across field pairs", fontsize=14)
    plt.xlabel("Field j"); plt.ylabel("Field i")
    plt.tight_layout()
    out = os.path.join(out_dir, "fca_heatmap_pairs.png")
    plt.savefig(out, dpi=300)
    plt.close()
    print(f"✅ saved {out}")
    return out

# ============ FIGURE 2: time trends for top FCA pairs (A) ============

def plot_top_pair_trends(out_dir, top_k=6):
    a_dir = find_latest_dir("field_convergence_A_")
    if not a_dir:
        print("⚠️ Part A output not found; skipping trends.")
        return []
    fca_path = os.path.join(a_dir, "fca_summary.csv")
    wy_path  = os.path.join(a_dir, "per_year_field_weights.csv")
    if not (os.path.exists(fca_path) and os.path.exists(wy_path)):
        print("⚠️ Missing Part A CSVs; skipping trends.")
        return []

    df_fca = pd.read_csv(fca_path)
    df_wy  = pd.read_csv(wy_path)
    df_fca = df_fca.sort_values("FCA_slope", ascending=False).head(top_k).copy()
    df_wy["pair"] = list(map(lambda xy: safe_pair(xy[0], xy[1]), zip(df_wy["field_i"], df_wy["field_j"])))

    outs = []
    for _, r in df_fca.iterrows():
        pair = safe_pair(r["field_i"], r["field_j"])
        sub = df_wy[df_wy["pair"] == pair].groupby("year", as_index=False)["weight_cosine"].mean().sort_values("year")
        if sub.empty: 
            continue
        plt.figure(figsize=(6.4,4.2))
        plt.plot(sub["year"], sub["weight_cosine"], marker="o")
        plt.grid(alpha=.3)
        plt.xlabel("Year"); plt.ylabel("Cosine weight")
        plt.title(f"Trend: {pair[0]} × {pair[1]}  (FCA={r['FCA_slope']:.4f})")
        plt.tight_layout()
        fn = f"pair_trend_{pair[0].replace('/','-')}__{pair[1].replace('/','-')}.png"
        out = os.path.join(out_dir, fn)
        plt.savefig(out, dpi=200); plt.close()
        outs.append(out)
    print(f"✅ saved {len(outs)} pair-trend figures")
    return outs

# ============ FIGURE 3: BES distribution (B) ============

def plot_bes_distribution(out_dir):
    b_dir = find_latest_dir("bridge_emergence_B_")
    if not b_dir:
        print("⚠️ Part B output not found; skipping BES distribution.")
        return None
    bes_path = os.path.join(b_dir, "bes_summary.csv")
    if not os.path.exists(bes_path):
        print("⚠️ bes_summary.csv missing; skipping BES distribution.")
        return None

    df = pd.read_csv(bes_path)
    x = pd.to_numeric(df["BES_delta"], errors="coerce").dropna().values
    plt.figure(figsize=(7,5))
    plt.hist(x, bins=50)
    plt.axvline(0, color="k", linewidth=1)
    plt.title("Distribution of Bridge Emergence Score (BESΔ)")
    plt.xlabel("BES (P_last − P_first)"); plt.ylabel("Authors")
    plt.tight_layout()
    out = os.path.join(out_dir, "bes_distribution.png")
    plt.savefig(out, dpi=200); plt.close()
    print(f"✅ saved {out}")
    return out

# ============ C helpers (scan articles if we need CPR distributions) ============

ART_DIR_TMPL = "articles_{year}_new"
ART_FILE = "all_articles_enhanced.jsonl"
YEARS = list(range(2018, 2026))

def stream_articles(year):
    path = os.path.join(ART_DIR_TMPL.format(year=year), ART_FILE)
    if not os.path.exists(path): 
        return
    with open(path, "r", encoding="utf-8") as f:
        for line in f:
            line=line.strip()
            if not line: continue
            try:
                yield json.loads(line)
            except Exception:
                continue

def rebuild_cpr_lists_from_articles(pairs):
    """Return dict: 'top'->list(cpr), 'bottom'->list(cpr), matching given pair sets"""
    pair_sets = {"top": set(pairs["top"]), "bottom": set(pairs["bottom"])}
    out = {"top": [], "bottom": []}
    for y in YEARS:
        for obj in stream_articles(y) or []:
            fields = obj.get("fields") or []
            fields = sorted(set(str(x).strip() for x in fields if str(x).strip()))
            if len(fields) < 2: continue
            paper_pairs = set(safe_pair(a,b) for a,b in itertools.combinations(fields,2))
            cpr = obj.get("citation_per_reference_ratio")
            if cpr is None:
                # fall-back
                ref = obj.get("referenced_works_count") or 1
                cited = obj.get("cited_by_count") or 0
                cpr = float(cited)/float(ref)
            for label in ("top","bottom"):
                if paper_pairs & pair_sets[label]:
                    out[label].append(float(cpr))
    return out

# ============ FIGS 4–6: from novelty_AB_* (C), with fallbacks ============

def plots_from_partC(out_dir, top_n_pairs=50):
    c_dir = find_latest_dir("novelty_AB_")
    a_dir = find_latest_dir("field_convergence_A_")
    made = []

    # 4) scatter FCA vs bridge share
    scatter_src = None
    if c_dir:
        scatter_src = os.path.join(c_dir, "plots", "scatter_fca_vs_bridge_share.png")
    if scatter_src and os.path.exists(scatter_src):
        out = os.path.join(out_dir, "scatter_fca_vs_bridge_share.png")
        import shutil; shutil.copy2(scatter_src, out); made.append(out)
        print(f"✅ copied {out}")
    else:
        # fallback: compute from link csv
        link_path = os.path.join(c_dir, "link_fca_bes_per_pair.csv") if c_dir else None
        if link_path and os.path.exists(link_path):
            df = pd.read_csv(link_path)
            plt.figure(figsize=(7,5))
            plt.scatter(df["FCA_slope"], df["share_bridge_authors"])
            plt.grid(alpha=.3); plt.xlabel("FCA_slope"); plt.ylabel("Bridge-author share")
            plt.title("FCA vs Bridge-author share (per pair)")
            plt.tight_layout()
            out = os.path.join(out_dir, "scatter_fca_vs_bridge_share.png")
            plt.savefig(out, dpi=180); plt.close(); made.append(out)
            print(f"✅ saved {out}")

    # 5) box of bridge-share (Top vs Bottom FCA)
    link_path = os.path.join(c_dir, "link_fca_bes_per_pair.csv") if c_dir else None
    if link_path and os.path.exists(link_path):
        df = pd.read_csv(link_path)
        df = df.dropna(subset=["FCA_slope","share_bridge_authors"])
        df = df.sort_values("FCA_slope", ascending=False)
        top = df.head(top_n_pairs)["share_bridge_authors"].values
        bot = df.tail(top_n_pairs)["share_bridge_authors"].values
        plt.figure(figsize=(6.6,4.8))
        plt.boxplot([top, bot], labels=["Top-FCA pairs","Bottom-FCA pairs"], showmeans=True)
        plt.ylabel("Bridge-author share"); plt.title("Bridge share: Top vs Bottom FCA")
        plt.grid(axis="y", alpha=.3); plt.tight_layout()
        out = os.path.join(out_dir, "box_bridge_share_top_vs_bottom.png")
        plt.savefig(out, dpi=180); plt.close(); made.append(out)
        print(f"✅ saved {out}")

    # 6) box CPR (Top vs Bottom FCA)
    # עדיף להעתיק אם קיים
    box_src = os.path.join(c_dir, "plots", "box_impact_top_vs_bottom.png") if c_dir else None
    if box_src and os.path.exists(box_src):
        out = os.path.join(out_dir, "box_cpr_top_vs_bottom.png")
        import shutil; shutil.copy2(box_src, out); made.append(out)
        print(f"✅ copied {out}")
    else:
        # fallback: נחשב רשימות CPR מחדש עבור Top/Bottom pairs
        if a_dir:
            fca_path = os.path.join(a_dir, "fca_summary.csv")
            if os.path.exists(fca_path):
                df_fca = pd.read_csv(fca_path).sort_values("FCA_slope", ascending=False)
                top_pairs = [safe_pair(i,j) for i,j in df_fca.head(top_n_pairs)[["field_i","field_j"]].values]
                bot_pairs = [safe_pair(i,j) for i,j in df_fca.tail(top_n_pairs)[["field_i","field_j"]].values]
                cpr_lists = rebuild_cpr_lists_from_articles({"top": top_pairs, "bottom": bot_pairs})
                plt.figure(figsize=(6.6,4.8))
                plt.boxplot([cpr_lists["top"], cpr_lists["bottom"]], labels=["Top-FCA pairs","Bottom-FCA pairs"], showmeans=True)
                plt.ylabel("Citations per Reference (CPR)"); plt.title("Impact: Top vs Bottom FCA")
                plt.grid(axis="y", alpha=.3); plt.tight_layout()
                out = os.path.join(out_dir, "box_cpr_top_vs_bottom.png")
                plt.savefig(out, dpi=180); plt.close(); made.append(out)
                print(f"✅ saved {out}")

    return made

# ============ FIGURE 7: p-values histogram (E) ============

def plot_pvals_histogram(out_dir):
    e_dir = find_latest_dir("e_significance_")
    if not e_dir:
        print("⚠️ Part E output not found; skipping p-values hist.")
        return None
    sig_path = os.path.join(e_dir, "fca_significance.csv")
    if not os.path.exists(sig_path):
        print("⚠️ fca_significance.csv missing; skipping p-values hist.")
        return None
    df = pd.read_csv(sig_path)
    p = pd.to_numeric(df.get("p_one_sided"), errors="coerce")
    q = pd.to_numeric(df.get("q_fdr"), errors="coerce")

    plt.figure(figsize=(12,4.6))
    plt.subplot(1,2,1)
    plt.hist(p.dropna(), bins=40)
    plt.title("Null-model p-values (one-sided)"); plt.xlabel("p"); plt.ylabel("#pairs")
    plt.subplot(1,2,2)
    plt.hist(q.dropna(), bins=40)
    plt.title("FDR q-values (BH)"); plt.xlabel("q"); plt.ylabel("#pairs")
    plt.tight_layout()
    out = os.path.join(out_dir, "pvals_histogram_E.png")
    plt.savefig(out, dpi=200); plt.close()
    print(f"✅ saved {out}")
    return out

# ============ main ============

def main():
    out_dir = ensure_outdir()
    print(f"\n📁 Output: {out_dir}\n")

    plot_fca_heatmap(out_dir)                 # Fig 1
    plot_top_pair_trends(out_dir, top_k=6)    # Fig 2 (6 דוגמאות)
    plot_bes_distribution(out_dir)            # Fig 3
    plots_from_partC(out_dir, top_n_pairs=50) # Figs 4–6
    plot_pvals_histogram(out_dir)             # Fig 7

    print("\n✅ All requested figures generated (where inputs existed).")

if __name__ == "__main__":
    main()
```
